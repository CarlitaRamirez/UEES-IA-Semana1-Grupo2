```python
# Importaciones estándar
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
```

# Configuración de visualización

```python
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

print("Entorno configurado correctamente en Google Colab")
```

```python
# Carga de datos de ejemplo (usaremos uno de Seaborn)
datos = sns.load_dataset('iris')
print(datos.head())
print(datos.info())
```

# Fundamentos de Matplotlib


La visualización de datos es una etapa clave del análisis de datos. Después de haber recopilado, almacenado y analizado datos, es esencial transformar los resultados de esos análisis en informes y visualizaciones gráficas.
Fuente: https://datascientest.com/es/todo-sobre-matplotlib

# 1.1 Gráfica

```python
# Gráfico Básico: Histograma
plt.figure(figsize=(10, 6))
plt.hist(datos['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Longitud del Sépalo')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

## Gráfico Scatter

```python
# Gráfico Avanzado: Scatter Plot con Subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Scatter Plot 1
ax[0].scatter(datos['sepal_length'], datos['petal_length'], c='green', alpha=0.6)
ax[0].set_title('Longitud de Sépalo vs. Pétalo')
ax[0].set_xlabel('Longitud del Sépalo')
ax[0].set_ylabel('Longitud del Pétalo')

# Scatter Plot 2 con diferenciación (ejemplo más avanzado)
for species in datos['species'].unique():
    subset = datos[datos['species'] == species]
    ax[1].scatter(subset['sepal_width'], subset['petal_width'], label=species, alpha=0.7)

ax[1].set_title('Ancho de Sépalo vs. Ancho de Pétalo por Especie')
ax[1].set_xlabel('Ancho del Sépalo')
ax[1].set_ylabel('Ancho del Pétalo')
ax[1].legend(title='Especie')

plt.tight_layout()
plt.show()
```

# Fundamentos de Seaborn

```python
# Gráfico Estadístico: Box Plot (comparación de distribuciones)
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=datos, palette='Set2')
plt.title('Distribución de la Longitud del Pétalo por Especie')
plt.ylabel('Longitud del Pétalo')
plt.xlabel('Especies')
plt.show()
```

```python
# Gráfico Avanzado: Pair Plot (matrices de dispersión)
# Advertencia: Esto puede tomar un momento y es un gráfico grande
sns.pairplot(datos, hue='species', markers=["o", "s", "D"], corner=True)
plt.suptitle('Pair Plot de las Características de Iris', y=1.02)
plt.show()
```

```python
# Gráfico de Distribución: Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_width', data=datos, inner='quartile', palette='pastel')
plt.title('Violin Plot del Ancho del Sépalo por Especie')
plt.show()
```

# Plotly

La biblioteca gráfica en Python de Plotly crea gráficos interactivos de calidad de publicación. Ejemplos de cómo hacer diagramas de líneas, diagramas de dispersión, gráficos de área, gráficos de barras, barras de error, diagramas de caja, histogramas, mapas de calor, subgráficos, ejes múltiples, gráficos polares y gráficos de burbujas.
Funte: https://plotly.com/python/

```python
# Gráfico Interactivo: Scatter Plot
fig = px.scatter(datos,
                 x="sepal_width",
                 y="sepal_length",
                 color="species",
                 title="Scatter Plot Interactivo de Iris",
                 hover_data=['petal_length', 'petal_width']) # Datos que aparecen al pasar el mouse

# Mostrar el gráfico (Plotly se renderiza directamente en Colab)
fig.show()

# Gráfico Interactivo: Box Plot
fig2 = px.box(datos,
              x="species",
              y="petal_length",
              color="species",
              title="Box Plot Interactivo de Longitud del Pétalo")
fig2.show()
```
